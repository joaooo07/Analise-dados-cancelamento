```python
import pandas as pd
df = pd.read_csv("/content/cancelamentos_sample.csv")
#df.pd-pd.read_excel(baseDeDados, sheet_name='nome_aba') <- le uma aba especifica
display(df)
```

* Cancelou será nossa variavel alvo (target) | Ela é categorica, logo faremos uma classificação<br>
vamos analisar se a base de dados está tratada.
<br> Um bom lugar para começar:
    * verificar informações irrelevantes
    * se existe linhas vazias
    * se está no formato certo

```python
#Verificando se existe informações irrelevantes
df.head()
```

```python
#Podemos verificar que "CustomerId" não é relevante para nossa analise. vamos remove-lo
df.drop(columns="CustomerID", inplace=True)
# O inplace=True Modifica o df original sem precisar atribuir a uma nova variavel
#outra forma de fazer seria
#df = df.drop(columns="CustomerID")
display(df)
```

```python
#Verificando valores nulos e se os dados estão no formato certo
df.info()
```

```python
#dropando os valores vazios
df.dropna(inplace=True)
df.info()
#Outras opções que temos para resolver esse problema, por exemplo quando você tem uma quantidade grande de valores não nulos
#é excluir a coluna inteira (caso nao impacte tanto na analise)
#ou preencher usando o comando pandas ffill e fillna
#fillna - preencher valores vazios
#ffill - preenche valores vazios com base no valor da linha anterior (funciona mais para dados com ordem cronologica)

```

```python
#Analisando o cancelamento dos clientes
#normalize true faz calcular a proporção
df['cancelou'].value_counts(normalize=True) * 100
#caso você queira formatar um numero
#df['cancelou'].value_counts(normalize=True).map('{:.1%}'.format)
```

```python
#Analise da causa dos cancelamentos
import plotly.express as px
for columns in df.columns:
  histogram = px.histogram(df, x='idade', color='cancelou', text_auto=True)
  histogram.show()
```

```python
# resolvendo alguns problemas para ver se a taxa de cancelamento melhora
# como fica minha base de dados se eu excluir todos os usuarios que ligaram mais de 4x pro callcenter?
df = df.loc[df['ligacoes_callcenter'] <= 4]
df['cancelou'].value_counts(normalize=True) * 100
```

```python
#excluindo o problema de dia de atraso
df = df.loc[df['dias_atraso'] <= 20]
df['cancelou'].value_counts(normalize=True) * 100
```

```python
#excluindo dados mensais
df = df.loc[df['duracao_contrato'] !="Monthly"]
df['cancelou'].value_counts(normalize=True) * 100
```
